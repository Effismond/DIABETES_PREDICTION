```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv("Diabetes.csv")
```

```python
print(df)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.corr()
```

```python
d=df.loc[(df["Glucose"]!=0)&(df["BloodPressure"]!=0)&(df["SkinThickness"]!=0)&(df["Insulin"]!=0)&(df["BMI"]!=0)]
```

```python
print(d.shape)
print(df.shape)
```

```python
d.describe()
```

```python
# df.replace(df["Glucose"]==0,value=d["Glucose"].mean(), inplace=True)
df["Glucose"].replace(0,d["Glucose"].mean(), inplace=True)
df["BloodPressure"].replace(0,d["BloodPressure"].mean(), inplace=True)
df["SkinThickness"].replace(0,d["SkinThickness"].mean(), inplace=True)
df["Insulin"].replace(0,d["Insulin"].mean(), inplace=True)
df["BMI"].replace(0,d["BMI"].mean(), inplace=True)
```

```python
df.describe()
```

```python
df.corr()
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(x_train, y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
print("predicted Values:")
print(y_pred)
print("Actual Values:")
print(y_test)
```

```python
y_test
```

```python
cf=confusion_matrix(y_test,y_pred)
print(cf)
print("classification Report for Testing Dataset:")
print(classification_report(y_test,y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNNclassifier=KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
KNNclassifier.fit(x_train, y_train)
```

```python
y_pred=KNNclassifier.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn import metrics
print("accuracy", metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
